This coincided with the rise of smartphones, which emphasized the need and competitiveness for more mobile-friendly versions of browsers. <a href="https://en.wikipedia.org/wiki/Browser_wars">[1]</a></p>
<p>In this notebook, we'll analyze the worldwide popularity of browsers over time using Google Trends. Although this won't give us direct market share figures, we can use Google Trends to get a sense of interest of a given browser over time and how that interest compares to other browsers. In particular, we will be looking at five major players over the past two decades: Mozilla Firefox (2002-), Apple's Safari (2002-), Google Chrome (2008-), Microsoft Internet Explorer (1995-2020), and Opera (1995-).</p>

  <p>Numbers represent search interest relative to the highest point on the chart for the given region and time. A value of 100 is the peak popularity for the term. A value of 50 means that the term is half as popular. A score of 0 means there was not enough data for this term.</p>

```python
# Ucitavanje podataka
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('datasets/worldwide_browser_trends.csv',parse_dates=['Month'])
df.set_index('Month',inplace=True)
df.head()
```

```python
#Izracunavanje rolling average - pitanje 1
rolling_six=df.rolling(window=6).mean()
rolling_six
rolling_six.plot(title='6. mesecni rolling avg')
```

```python
#Izracunavanje pct_change

pct_change_quarterly=df.pct_change(3)*100
pct_change_quarterly=pct_change_quarterly[pct_change_quarterly.index.year>=2009]
pct_change_quarterly
pct_change_quarterly.plot(title='Kvartalni PCT',subplots=True,figsize=(20,10))
```

```python
#Izracunavanje Google Trendsa

google=df[df.index.year.isin([2009,2012,2015,2018])]
chrome_trends=google.pivot_table(values='Google Chrome',columns=google.index.year,index=google.index.month)
chrome_trends
chrome_trends.plot(title='Google Chrome performanse pretrage u 2009. ,2012. ,2025. i 2018. godini',figsize=(12,8))
```
