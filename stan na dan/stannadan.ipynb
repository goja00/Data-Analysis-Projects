```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd

def brisanje_reda(konacan,a):
    for i in a:
       konacan.append(i.strip('\n'))

def brisanje_razmaka(a):
    x=list()
    for i in a:
        x.append(" ".join(i.split()))

    return x
```

```python

re=requests.get('https://www.apartmani-u-beogradu.com/stan-na-dan-novi-sad?long=19.7794011&lat=45.2715508&zoom=13&instant=0&city_id%5B%5D=5&page=6')
df=bs(re.content,'html.parser')
apartmani=df.find_all(class_='singleApartmentText')
cene=df.find_all(class_='listing box-shadow')

naslov=list()
adresa=list()
info=list()
cena=list()

for a in apartmani:
    naslov.append(a.find(class_='singleApartmentName').get_text())
    adresa.append(a.find(class_='singleApartmentLocation').get_text())
    info.append(a.find(class_='singleApartmentInfo').get_text())

for c in cene:
    cena.append(c.find(class_='singleApartmentPrice').get_text())


naslov1=naslov[:18]
adresa1=adresa[:18]
info1=info[:18]
cena1=cena[:18]

konacan_naslov=list()
konacna_adresa=list()
konacan_info=list()
konacna_cena=list()

brisanje_reda(konacan_naslov,naslov1)
brisanje_reda(konacna_cena,cena1)
konacna_adresa=brisanje_razmaka(adresa1)
konacan_info=brisanje_razmaka(info1)
konacna_cena=brisanje_razmaka(konacna_cena)
konacna_cena1=list()
for k in konacna_cena:
    konacna_cena1.append(k.split()[1])

kvadratura=list()
struktura=list()
osobe=list()
ocena=list()

for x in konacan_info:
    t=x.split()
    kvadratura.append(t[0] + t[1])
    struktura.append(t[2])
    osobe.append(t[3] + " " + t[4])
    ocena.append(t[5])
    
```

```python
recnik={'Naziv':konacan_naslov,'Adresa':konacna_adresa,'Kvadratura':kvadratura,'Struktura':struktura,'Osobe':osobe,'Ocena':ocena,'Cena u €':konacna_cena1}
```
